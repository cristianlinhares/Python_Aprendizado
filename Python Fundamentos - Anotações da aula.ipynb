#  Operadores lógicos:

# Tipo dicionário: 

```python
#Enquanto as listas são estruturas indexadas, o dicionário segue a estrutura de chave-valor (no preview do vs code essa diferença fica bem clara)


#Sintaxe dicionário: {}

pessoa = {'nome ':'Prof(a) Ana', 'idade ': 38, 'cursos ': ['Inglês', 'Português']}
print(type(pessoa))

print(len(pessoa)) #no caso do dicionário, retorna a quantidade de PARES chave-valor que existem no dicionário

print(pessoa['nome ']) #note que para acessar determinado valor numa lista buscamos o índice do item, no dicionário buscamos sua chave para que ele retorne o valor
print(pessoa['cursos '][1]) #acessa um item da lista que está dentro do dicionário

print(pessoa.keys()) #chama as chaves do dicionário
print(pessoa.values()) #chama os valores do dicionário
print(pessoa.items())

print(pessoa.get('idade '))
print(pessoa.get('jonas', 'não há esse valor no dicionário')) #configuração para retornar um conteúdo qualquer caso não localize o requerido no dicionário
```

```python
pessoa = {'nome': 'Prof Alberto','idade':43, 'cursos': ['React', 'Python']}
pessoa['idade'] = 44 #estou substituindo o item idade
pessoa['cursos'].append("Angular") #estou incluindo um item

print(pessoa)

print(pessoa.pop('idade')) #tanto retorna o valor requerido quanto retira ele de vez do dicionário
print(pessoa)

pessoa.update({'idade':40, 'sexo':'M'})
print(pessoa)

del pessoa['cursos']
print(pessoa)

pessoa.clear() #limpa todo dicionário
print(pessoa)
```

# Tipo Conjunto(Set): 

```python
#conjunto não aceita repetição e não é uma estrutura indexada
#sintaxe = {}

a = {1, 2, 3}
print( 1 in a, 4 not in a)

print( a == {2, 3, 1, 1, 1}) #O resultado é true porque em conjuntos:a ordem não importa e os elementos duplicados serão removidos

'''Operações'''
c1 = {1, 2}
c2 = {2, 3}

print(c1.union(c2))
print(c1.intersection(c2))
print(c1.update(c2)) #altera o conteúdo de c1 incluindo os elementos de c2, note que essa operação na verdade é uma união, mas uma união que altera o valor de c1 de modo irreversível
print(c1)
print(c2<=c1) #c2 está contido em c1? 

'''-------------'''

print({1,2,3} - {2}) #diferença entre os conjuntos
print(c1-c2)
```

# Interpolação: 

```python
#interpolar = substituir valores dentro da string

nome, idade = 'Ana', 30 #só pra mostrar que é possível criar duas variáveis na mesma linha

''' 1º MÉTODO (versão mais antiga e menos recomendada): O % é um método antigo de interpolação assim, 
usa-se %s para substituir string (por isso o s), %d para substituir números inteiros e %f para substituir float. VEJA:'''

print('Nome: %s Idade: %d' % (nome,idade)) 

'''2º MÉTODO: '''

print('Nome: {0} Idade: {1}'.format(nome,idade))

'''3º MÉTODO(mais atual, implementada depois do python 3.6, e recomendada) '''

print(f'Nome: {nome} Idade: {idade}') #método fstring

'''4º MÉTODO: '''

from string import Template
s = Template('Nome: $nome Idade: $idade')
print(s.substitute(nome=nome, idade=idade))
```
